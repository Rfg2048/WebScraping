```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import time
import datetime
import smtplib
import csv
```

```python
#Connecting to DID website to get the html information

url = "https://www.did.ie/products/samsung-50-cu7100-uhd-led-4k-hdr-smart-tv-black-ue50cu7100kxxu"

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(url, headers = headers)

soup = BeautifulSoup(page.content, "html.parser")

#Extracting title of product and proce from the webpage

title = soup.find(class_="product_name title").get_text()
price = soup.find(class_="span_money money sale").get_text()

#Formating price and title values to remove the Euro sign from price and empty spaces

price = price.strip()[1:]
title = title.strip()

#Variable to get current date

today = datetime.date.today()

print(title)
print(price)
print(today)
```

```python
# Header and data variables to be included in the csv

header = ["Product Title", "Product Price","Date"]
data = [title,price,today]

#Creation of CSV file writing the header and first row of data

with open("DIDScrape.csv", "w", newline = "", encoding = "UTF8") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```

```python
#Putting code in a class to be called with a timer

def check_price():
    import datetime
    import csv
    
    url = "https://www.did.ie/products/samsung-50-cu7100-uhd-led-4k-hdr-smart-tv-black-ue50cu7100kxxu"

    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    page = requests.get(url, headers = headers)

    soup = BeautifulSoup(page.content, "html.parser")
    
    title = soup.find(class_="product_name title").get_text()
    price = soup.find(class_="span_money money sale").get_text()
    
    price = price.strip()[1:]
    title = title.strip()
    
    today = datetime.date.today()
    
    header = ["Product Title", "Product Price","Date"]
    data = [title,price,today]
    
# Instead of writing in the csv file now using "a+" the new row will be appended in file
    
    with open("DIDScrape.csv", "a+", newline = "", encoding = "UTF8") as f:
        writer = csv.writer(f)
        writer.writerow(data)
```

```python
#Timer to run the code every minute (60 seconds)

while(True):
    check_price()
    time.sleep(60)
```
